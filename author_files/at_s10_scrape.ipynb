```python
import requests
from bs4 import BeautifulSoup as BS
import lxml
import re
import pandas as pd
import csv
```

```python
# create list of episode titles from season 2

s10 = [
    "The Wild Hunt",
    "Always BMO Closing",
    "Son of Rap Bear",
    "Bonnibel Bubblegum",
    "Seventeen",
    "Ring of Fire",
    "Marcy & Hunson",
    "The First Investigation",
    "Blenanas",
    "Jake the Starchild",
    "Temple of Mars",
    "Gumbaldia",
    "Come Along With Me",
]

# format episode titles for use in url

season_10 = []

for i in s10:
    if ' ' in i:
        i = i.replace(' ', '_')
        season_10.append(i)
    else:
        season_10.append(i)
```

```python
# fetch trancripts for each episode and add to single list
# NOTE: this Community content is available under CC-BY-SA unless otherwise noted.

s10_trans = []

for ep in season_10:
    url = f'https://adventuretime.fandom.com/wiki/{ep}/Transcript'
    site = requests.get(url)
    source_code = site.content
    soup = BS(source_code, 'lxml')
    transcript = soup.find_all('dd')
    for i in transcript:
        s10_trans.append(i)
```

```python
# convert b4.element.tag to strings

s10_trans_str = []

for i in s10_trans:
    string = str(i)
    s10_trans_str.append(string)
```

```python
# remove superflous stuff

s10_sentences = []

for x in s10_trans_str:
    x = x.replace(']', '')
    x = x.replace('[', '')
    x = x.replace('(', '')
    x = x.replace(')', '')
    x = re.sub('<.+?>', '', x)
    x = x.replace(':', ' says') # the ':' signifies a character line
    s10_sentences.append(x)
```

```python
df = pd.DataFrame({
    'text' : s10_sentences
})

df.to_csv('at_s10_text.csv')
```
